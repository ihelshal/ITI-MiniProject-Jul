```python
import numpy as np
```

```python
x = np.array([0, 1, 30 , 90])

print(np.sin(x))
print(np.cos(x))
print(np.tan(x))
```

```python
#inserve sin
print(np.arcsin(x)) 
```

```python
#convert angles from radians to degress
x = np.array([-np.pi, -np.pi/2, np.pi/2, np.pi])

print(np.degrees(x))
print(np.rad2deg(x))
```

```python
#convert angles from degress to radians
x = np.array([-180., -90.,   90.,  180.])

print(np.radians(x))
print(np.deg2rad(x))
```

```python
#HyperBolic Function

x = np.array([-1, 0 , 1])

print(np.sinh(x))
print(np.cosh(x))
print(np.tanh(x))
```

```python
x = np.array([-1, 0 , 1, 3.32222,4.433333,5.43333])

print(np.round(x, 3))
print(np.floor(x))
print(np.ceil(x))
```

```python
x = np.array([-1, 0 , 1, 3.32222,4.433333,5.43333])

# e^x
np.exp(x)
```

```python
np.exp2(x)
```

```python
x = np.array([1, np.e, np.e**2])

print(np.log(x))
print(np.log10(x))
print(np.log2(x))
```

```python
x = np.array([3,4,5])
y = np.array([1,0,-1])

np.add(x, y)
```

```python
np.multiply(x, y)
```

```python
x = np.array([1+2j, 3+4j, 5+6j])

real = x.real
img = x.imag

print(real, img)
```

```python
np.conjugate(x[0])
```
