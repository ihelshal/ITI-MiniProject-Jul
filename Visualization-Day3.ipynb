```python
import seaborn as sns

import warnings
warnings.filterwarnings('ignore')
```

```python
tips = sns.load_dataset('tips')
tips
```

```python
sns.relplot(
    data=tips,
    x='total_bill',
    y='tip',
    col='time',
    size='size',
    style='sex',
    hue='sex'
)
```

```python
dots = sns.load_dataset('dots')
dots
```

```python
dots['align'].unique()
```

```python
sns.relplot(
    data=dots,
    kind='line',
    x='time',
    y='firing_rate',
    col='align',
    size='coherence',
    style='choice',
    hue='choice'

)
```

```python
df = sns.load_dataset('fmri')
df
```

```python
sns.relplot(
    data=df,
    kind='line',
    x='timepoint',
    y='signal',
    hue='event',
    style='event',
    col='region'

)
```

```python
sns.lmplot(
    data=tips,
    x='total_bill',
    y='tip',
    col='time',
    hue='sex'
)
```

```python
sns.displot(
    data=tips,
    x='total_bill',
    col='time', 
    kde=True
)
```

```python
sns.catplot(
    data=tips,
    kind='swarm',
    x='day',
    y='total_bill',
    hue='sex'
)
```

```python
sns.catplot(
    data=tips,
    kind='bar',
    x='day',
    y='total_bill',
    hue='sex'
)
```
